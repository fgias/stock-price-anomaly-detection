```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import pandas as pd

# Assuming x is your dataframe
# x = pd.DataFrame({
#     'col3': [..], 'col4': [..], 'col5': [..], 'col6': [..]
# })

# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Set font sizes
title_fontsize = 22
label_fontsize = 22
tick_fontsize = 20
inset_title_fontsize = 20
inset_label_fontsize = 20
inset_tick_fontsize = 20

# Plot each column in a different subplot
axs[0, 0].plot(x.iloc[:, 9])
axs[0, 0].set_title('Discriminator Loss', fontsize=title_fontsize)

# Add inset to the first plot (axs[0, 0]) slightly higher in the lower right
inset_ax1 = inset_axes(axs[0, 0], width="100%", height="100%",
                       bbox_to_anchor=(0.45, 0.1, 0.4, 0.3), 
                       bbox_transform=axs[0, 0].transAxes, loc='lower right')
inset_ax1.plot(x.iloc[:500, 9])
# inset_ax1.set_xlabel('Training Step', fontsize=inset_label_fontsize)
# inset_ax1.set_ylabel('Value', fontsize=inset_label_fontsize)
inset_ax1.set_title('First 500 Steps', fontsize=inset_title_fontsize)
inset_ax1.tick_params(axis='both', which='major', labelsize=inset_tick_fontsize)

axs[0, 1].plot(x.iloc[:, 12].apply(lambda x: float(x[:4])))
axs[0, 1].set_title('Discriminator Accuracy', fontsize=title_fontsize)

axs[1, 0].plot(x.iloc[:, 15])
axs[1, 0].set_title('Generator MSE', fontsize=title_fontsize)

# Add inset to the third plot (axs[1, 0])
inset_ax2 = inset_axes(axs[1, 0], width="100%", height="100%",
                       bbox_to_anchor=(0.45, -.95, 0.4, 0.3), 
                       bbox_transform=axs[0, 0].transAxes, loc='lower right')
inset_ax2.plot(x.iloc[:500, 15])
# inset_ax2.set_xlabel('Training Step', fontsize=inset_label_fontsize)
# inset_ax2.set_ylabel('Value', fontsize=inset_label_fontsize)
inset_ax2.set_title('First 500 Steps', fontsize=inset_title_fontsize)
inset_ax2.tick_params(axis='both', which='major', labelsize=inset_tick_fontsize)

axs[1, 1].plot(x.iloc[:, 18].apply(lambda x: float(x[:4])))
axs[1, 1].set_title('Generator Binary Loss', fontsize=title_fontsize)

# Set labels for each subplot
for ax in axs.flat:
    ax.set(xlabel='Training Step', ylabel='Value')
    ax.xaxis.label.set_size(label_fontsize)
    ax.yaxis.label.set_size(label_fontsize)
    ax.tick_params(axis='both', which='major', labelsize=tick_fontsize)

# Adjust layout to prevent overlap
fig.tight_layout()

# Display the plot
plt.savefig("training-plots.pdf")
```
