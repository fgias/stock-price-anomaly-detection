## Data cleaning

```python
df = pd.DataFrame(Y, columns=['density'])
df.index = pd.Series(X)
df.plot()
```

```python
window = 10
df['density_average'] = None
for i in df.index:
    df.loc[i, 'density_average'] = df.loc[i - window:i, 'density'].sum()/window

df
```

## GNN data

```python
cams = pd.read_csv("traffic_data/dataset1/CamerasGPSinfo.csv")
```

```python
raw = pd.read_pickle('traffic_data/dataset1/dens_0201')
raw
```

```python
date = "0201"
cam_id = 99
name = date + "Cam" + str(cam_id)
# Step 1: Filter out rows with non-iterable entries or incorrect lengths
filtered = raw[raw[name].apply(lambda x: isinstance(x, (list, tuple)) and len(x) == 2)]

# Step 2: Further filter to remove specific invalid Y values (e.g., 1)
valid = filtered[filtered[name].apply(lambda x: x[1] != 1)]

# Extract X and Y using list comprehension, safely assuming all remaining rows are proper lists or tuples
X = [item[0] for item in valid[name]]
Y = [item[1] for item in valid[name]]

df = pd.DataFrame([])
df["Time"] = X
df["Time"] = df["Time"].apply(lambda x: int(x[0:2])*60 + int(x[2:4]))
df["Density"] = Y
df["CamId"] = int(cam_id)
```

```python
df.merge(cams, on=['CamId'])
```

```python
row_num = 0
# Step 1: Filter out rows with non-iterable entries or incorrect lengths
filtered = raw.iloc[row_num][raw.iloc[row_num].apply(lambda x: isinstance(x, (list, tuple)) and len(x) == 2)]

# Step 2: Further filter to remove specific invalid Y values (e.g., 1)
valid = filtered[filtered.apply(lambda x: x[1] != 1)]

gnn_data = pd.DataFrame(valid)
```

```python
gnn_data["CamId"] = gnn_data.index.map(lambda x: int(x[x.find('Cam') + 3:]))
gnn_data["Date"] = gnn_data.index.map(lambda x: x[0:4])
gnn_data["Time"] = gnn_data[row_num].apply(lambda x: int(x[0][0:2])*60 + int(x[0][2:4]))
gnn_data["Density"] = gnn_data[row_num].apply(lambda x: x[1])
gnn_data.drop(row_num, axis=1, inplace=True)
gnn_data.reset_index(drop=True, inplace=True)
```

```python
cams.drop_duplicates('CamId', inplace=True)
```
