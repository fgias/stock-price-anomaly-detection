# Parsing the GPS data

```python
import pandas as pd
import re
```

```python
file = open("traffic_data/dataset1/CamerasGPSinfo.txt", "r")
lines = file.readlines()
```

```python
ids = []
longs = []
lats = []
bearings = []

id_pattern = r"<CameraDevice><Id>(\d+)</Id>"
coords_pattern = r"POINT \(([-+]?\d*\.\d+)\s+([-+]?\d*\.\d+)\)"
bearing_pattern = r"<Bearing>(\d+)</Bearing>"
for line in lines:
    matches_id = re.findall(id_pattern, line)
    matches_coords = re.findall(coords_pattern, line)
    matches_bearing = re.findall(bearing_pattern, line)
    if matches_id and matches_coords:
        id = matches_id[0]
        long, lat = matches_coords[0]
        bearing = matches_bearing[0]
        ids.append(id)
        longs.append(long)
        lats.append(lat)
        bearings.append(bearing)
```

```python
cameras = pd.DataFrame([])
cameras['CamId'] = ids
cameras['Longitude'] = longs
cameras['Latitude'] = lats
cameras['Bearing'] = bearings
```

```python
cameras.to_csv("traffic_data/dataset1/CamerasGPSinfo.csv", index=False)
```
